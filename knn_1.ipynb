```python
import math
```

```python
def euclideanDistance(instance1, instance2, length):
    distance = 0 
    for x in range(length):
        distance += pow((instance1[x]-instance2[x]), 2)
    return math.sqrt(distance)
```

```python
data1 = arr[0]
data2 = arr[2888]
distance = euclideanDistance(data1, data2, cols-1)
print("Distance: ", distance)
```

```python
import operator
def getNeighbours(trainingSet, testInstance, k):
    distances = []
    for x in range(len(trainingSet)):
        dist = euclideanDistance(testInstance, trainingSet[x], cols-1)
        distances.append((trainingSet[x], dist))
    distances.sort(key=operator.itemgetter(1))
    neighbours = []
    for x in range(k):
        neighbours.append(distances[x][0])
    return neighbours
```

```python
trainSet = trainingSet
testInstance = testSet[0][:-1]
k = 3
neighbours = getNeighbours(trainSet, testInstance, k)
print(neighbours)
```

```python
def getResponse(neighbours):
    classVotes = {}
    for x in range(len(neighbours)):
        response = neighbours[x][-1]
        if response in classVotes:
            classVotes[response] += 1
        else:
            classVotes[response] = 1
    sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)
    return sortedVotes[0][0]
```

```python
print(getResponse(neighbours))
```

```python
def getAccuracy(testSet, predictions):
    correct = 0
    for x in range(len(testSet)):
        if testSet[x][-1] == predictions[x]:
            correct += 1
    return (correct/float(len(testSet))) * 100.0
```

```python
predictions = []
k = 3
for x in range(len(testSet)):
    neighbours = getNeighbours(trainingSet, testSet[x], k)
    result = getResponse(neighbours)
    predictions.append(result)
    print("predicted = ", result, "actual = ",testSet[x][-1])
accuracy = getAccuracy(testSet, predictions)
print("Accuracy : ", accuracy, "%")
```

```python
actual = [testSet[x][-1] for x in range(len(testSet))]
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(actual,predictions))
print(classification_report(actual,predictions))
```
